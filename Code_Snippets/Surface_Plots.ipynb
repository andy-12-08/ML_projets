```python
###Bonus Question

import matplotlib.pyplot as plt
x = np.linspace(-10,10,150)
y = np.linspace(-10,10,150)
X,Y = np.meshgrid(x,y)
Z = np.cos(X)*np.cos(Y)*np.exp(-(X/5)**2-(Y/5)**2)

extent = np.min(x), np.max(x), np.min(y), np.max(y)


plt.figure(figsize=(10,8))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.5, hspace=0.5)

plt.subplot(2,2,1)
plt.title('Plot 1')
plt.pcolor(Z)
plt.xlabel("X")
plt.ylabel("Y")
cbar = plt.colorbar()
cbar.set_label('Color Intensity')


plt.subplot(2,2,2)
plt.title('Plot 2')
plt.contour(Z, extent = extent)
cbar = plt.colorbar()
cbar.set_label('Color Intensity')


plt.subplot(2,2,3)
plt.title('Plot 3')
plt.contourf(Z, extent=extent)
plt.xlabel("X")
plt.ylabel("Y")
cbar = plt.colorbar()
cbar.set_label('Color Intensity')


plt.subplot(2,2,4)
plt.title('Plot 4')
plt.imshow(Z, extent=extent)
plt.xlabel("X")
cbar = plt.colorbar()
cbar.set_label('Color Intensity')

plt.set_cmap("gist_rainbow")
plt.savefig('Plots.png')
plt.show()

```
